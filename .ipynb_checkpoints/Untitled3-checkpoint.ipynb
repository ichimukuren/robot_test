```python
import numpy as np
from numpy.linalg import inv
X = np.array([[1,50],[1,60],[1,70],[1,100]])
```

```python
Y = np.array([[10],[30],[40],[50]])
```

```python
X_Y_X = (X.T).dot(X)
```

```python
X_Y_X_inv = inv(X_Y_X)
w = X_Y_X_inv.dot(X.T).dot(Y)
```

```python
x_min  = X[:,1].min()
x_max  = X[:,1].max()
```

```python
X_sup = np.linspace(x_min,x_max,num=100)
```

```python
Y_model = w[0][0] + w[1][0]*X_sup
```

```python
plt.scatter(X[:,1],Y[:,0],100,'b','o',alpha=0.6)
plt.plot(X_sup,Y_model)
plt.show()
```

```python
import random

import numpy as np

import matplotlib.pyplot as plt

  
# Вероятность двигаться вверх или вниз

prob = [0.05, 0.95]  

  
# статически определяя начальную позицию

start = 2  

positions = [start]

  
# создание случайных точек

rr = np.random.random(100)

downp = rr < prob[0]

upp = rr > prob[1]


  

for idownp, iupp in zip(downp, upp):

    down = idownp and positions[-1] > 1

    up = iupp and positions[-1] < 4
    

    positions.append(positions[-1] - down + up)

  
# построение графика случайного блуждания в 1D
plt.plot(positions)
plt.show()
```

```python
import numpy

import pylab

import random

  
# определение количества шагов

n = 10000

  
# создание двух массивов для хранения координат x и y
#of size равно количеству size и заполнено нулями

x = numpy.zeros(n)

y = numpy.zeros(n)

  
# заполнение координат случайными величинами

for i in range(1, n):

    val = random.randint(1, 4)

    if val == 1:

        x[i] = x[i - 1] + 1

        y[i] = y[i - 1]

    elif val == 2:

        x[i] = x[i - 1] - 1

        y[i] = y[i - 1]

    elif val == 3:

        x[i] = x[i - 1]

        y[i] = y[i - 1] + 1

    else:

        x[i] = x[i - 1]

        y[i] = y[i - 1] - 1

      

  
# заговоры:

pylab.title("Random Walk ($n = " + str(n) + "$ steps)")

pylab.plot(x, y)

pylab.savefig("rand_walk"+str(n)+".png",bbox_inches="tight",dpi=600)

pylab.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
def laplacian_mod(img,threshold=[200,255]):
    kernal = np.int0([
        [ 0, 2, 0],
        [ 2,-5, 2],
        [ 0, 2, 0]
    ])
    
    if img.ndim == 3:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    return cv2.threshold(cv2.filter2D(img,-1,kernal),threshold[0],threshold[1],0)[1]
```
